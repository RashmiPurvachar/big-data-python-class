# Tensor Flow

**Machine learning (ML)** is the study of algorithms and mathematical models that computer systems use to progressively improve their performance on a specific task.

Machine learning algorithms build a mathematical model of sample data, known as "training data", in order to make predictions or decisions without being explicitly programmed to perform the task.

**Major categories of ML**

- Supervised Learning: The correct classes of the training data are known.

- Unsupervised Learning: The correct classes of the training data are not known.

- Reinforcement Learning: Allows the machine or software agent to learn its behavior based on feedback from the environment. This behavior can be learnt once and for all, or keep on adapting as time goes by.

**Machine Learning Techniques**

1. Classification: is a kind of problem wherein the outputs are categorical in nature, like Yes or No, True or False, 0 or 1. 
    - This predicts classes from observations.
    - Classification algorithms: KNN, Naive Bayes, Decision Tree and Random Forest.


2. Clustering: is the task of grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar to each other. 

    - Group observations into “meaningful” groups.
    - Clustering algorithms: K-Means, Hierarchical clustering
    
    
3. Regression(prediction): is a measure of the relation between the mean value of one variable (e.g. output) and corresponding values of other variables (e.g. time and cost). 
    - Regression analysis is a statistical process for estimating the relationships among variables.
    - Regression means to predict the output value using training data.
    - Predict value from observations.
    - Regression algorithms: Logistic regression, Linear and Polynomial Regression, Regression Trees.

**Deep Learning:** 

It is a subset of ML and it is inspired by the structure and functioning of a human brain.

Unlike ML the data is more complicated and and unstrucuted like an image or video files. One of the core components in deep learning is **Neural network**. A neural network has one to many hidden layers between an input and an output layer. 
Anything more than one hidden layer is known as Deep Neural Network. 

Deep Learning learns from the data that is unstructured and uses complex algorithms to train a neural net.

Input Layer - It accepts large amounts of data as an input to build the data. (Images, Pixel values of images, Video files etc.,). This data will be passed onto the hidden layers.

Hidden layer - These process the data by performing complex operations and computations and carries out feature extraction and then passes the vlaues to the output layer.

Output layer - This generates the predicted output by applying suitable activation functions.

### Tensor Flow

Tensor flow is one of the top Deep Learning libraries, open source, developed by Google Brain Team and written in Python, C++ and CUDA. 

TensorFlow bundles together a slew of machine learning and deep learning/neural networking models and algorithms and makes them useful by way of a common metaphor.

- This was developed originally to run large numerical computations.

- This accepts data in the form of multidimentional arrays of higher dimentions called Tensors.

- It works on the basis of Data FLow Graphs that have nodes and edges, while nodes represent mathematical operations,  the edges represent the data.

**Tensor** is a generalization of vectors and matrices of potentially higher dimensions. Arrays of data with differnent dimensions and ranks that are fed as input to the neural network called Tensors.

Tensors offer a structured and compact way of handling and storing the complicated data during computataion. 

**Dimension:** It is the number of elements which takes the form of a dimension.  

**Ranks:** It is traditionally a dimension itself which gives us the value of level of dimensions/array. 

For instance,
- s = [100] is s tensor of Rank 0, as it has a single element. (Also called as Scalar) 
- v = [1,2,3] has a rank 1, this is one-dimentional array having a row of elements, so this is a vector. 
- m = [1,2,3][4,5,6] has a rank 2 and 2-dimentional and hence is also a matrix.

### Data Flow Graph

Computation on data needs to be done only when the data is in place and is stored in a tensor, this computation  happens in a form of a 'Graphs'. Unlike traditional programming, graphs and various nodes has to be prepared which are executed in a form of a session using the data from the tensors. 

Each of these computation is represented in a Tensor flow as a Data Flow Graph.

A TensorFlow object creation in a will provide a default graph but, multiple graphs can be created as and when needed. 

The graph gets executed and processes all the input data (Data is fed in the form of 'Placeholders').

Each node in the graph represents a mathematical operation and each edge represents a multidimentional arrays(tensors).

Computational graph is the graph of programming logic which TensorFlow builds in the memory.

Once the graph(s) is/are available execution can be enabled either on a regular CPU or on GPUs in a distributed mode so that the process gets much faster. It also enables creating large scale neural networks as computing can be distributed across several CPUs and GPUs. 



### Program elements in TensorFlow

TensorFlow programs work on two basic concepts.,

1. Building a computational graph (Prepare a graph and create a session)
2. Executing a computational graph (in the session created)

**Constants** are parameters whose value does not change. To define a constant we use tf.constant() command. Values set, cannot be changed at any point of time during the execution.

Example: 

    a =  tf.constant(2.0, tf.float(32)) // specifying along with type
    b = tf.constant(3.0)
    print(a, b)

**Variables** allow us to add new trainable parameters to graphs. To define a variable we use tf.Variable() command and initialize them before running the graph in a session. 

Example:

    W = tf.Variable([0.3],dtype=f.float32)
    b = tf.Variable([- 0.3],dtype=f.float32)
    x = tf.placeholder(tf.float32)
    linear_model = W*x+b

**Placeholders** allow us to feed data to TensorFlow model from outside a model(Primary purpose, however it allows data to be fed within a model too). It permits to assign a value later. To define a placeholder we use tf.placeholder() command.

feed_dict is a parameter to populate the placeholders and specifies tensors that provides concrete values to placeholders

Example:

    a = tf.placeholder(tf.float32)
    b = a*2
    with tf.session() as sess:
    result = sess.run(b, feed_dict={a:3.0})
    print result
    
**Session** is created and run to evaluate the nodes. This is called as a tensorFlow runtime.

Example:

    a = tf.constant(5.0)
    b = tf.constant(3.0)
    c = a*b
    
    # To launch a session
    sess = tf.session()
    
    # To evaluate the tensor c
    print(sess.run(c))

**Importing necessary libraries and the dataset**

```python
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
import numpy as np

from numpy import genfromtxt
from sklearn.datasets import load_boston
```

**Creating methods for data preparation**

To read the text from a .csv file to a Numpy array we define a method: 'read_txt_from_dataset' 

```python
def read_txt_from_dataset(filePath,delimiter=','):
    return genfromtxt(filePath, delimiter=delimiter)
```

To read the Boston dataset from scikit-learn library we define a method: read_boston_dataset. 
Return both data features and labels separately. 

```python
def read_boston_dataset():
    boston = load_boston()
    features = np.array(boston.data)
    labels = np.array(boston.target)
    return features, labels
```

To normalize the features of the dataset we define a method: 'normalize_features'. 

```python
def normalize_features(dataset):
    m = np.mean(dataset,axis=0)
    sigma = np.std(dataset,axis=0)
    return (dataset - m)/sigma
```

To append the bias term to the normalized features we define method: 'append_bias'. 
This method will reshape the features in a proper format that fits and can be fed to TensorFlow library. 

```python
def append_bias(features,labels):
    n_training_samples = features.shape[0]
    n_dim = features.shape[1]
    x = np.reshape(np.c_[np.ones(n_training_samples),features],[n_training_samples,n_dim + 1])
    y = np.reshape(labels,[n_training_samples,1])
    return x, y
```

```python
learning_rate = 0.01
training_epochs = 1000

# To return a new array of given shape and type, without initializing entries.
house_price = np.empty(shape=[1],dtype=float) 

X = tf.placeholder(tf.float32,[None,dimension])
Y = tf.placeholder(tf.float32,[None,1])
W = tf.Variable(tf.ones([dimension,1]))

init = tf.global_variables_initializer()
```

**Linear regression in TensorFlow**

tf.train.GradientDescentOptimizer() is an optimizer that implements the gradient descent algorithm.

[https://www.tensorflow.org/api_docs/python/tf/train/GradientDescentOptimizer]

```python
# To multiply features matrix with weights matrix. This will be used for Prediction
y_ = tf.matmul(X, W)

# Computes the mean of elements across dimensions of a tensor. This in turn computes cost or loss. 
cost = tf.reduce_mean(tf.square(y_ - Y))

# To minimize the cost function above
training_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

**Training Regression Model**

```python
# Create and initiate a new session
sess = tf.Session()
sess.run(init)

for epoch in range(training_epochs):
    sess.run(training_step,feed_dict={X:train_x,Y:train_y})
    house_price = np.append(house_price,sess.run(cost,feed_dict={X: train_x,Y: train_y}))
```

To plot the cost as a function of number of iterations using Matplotlib

```python
plt.plot(range(len(house_price)),house_price)
plt.axis([0,training_epochs,0,np.max(house_price)])
plt.show()
```

To make predictions on test dataset and calculate mean square error. 

```python
pred_y = sess.run(y_, feed_dict={X: test_x})
mse = tf.reduce_mean(tf.square(pred_y - test_y))
print("Mean Square Error: %.4f" % sess.run(mse)) 
```
